**Grafico de dispersion:** Utilizamos este tipo de graficos para comparar los valores que toman dos variables distintas, una de las cuales se representa a lo largo del eje x y la otra a lo largo del eje y. La grafica resultante nos permite identificar visualmente la posible correlacion entre las dos variables

```python
# Matplotlib scatterplot o diagrama de dispersion
from matplotlib import pyplot as plt #Importamos la libreria

#Puntos de datos
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [125, 32, 54, 253, 67, 87, 233, 56, 67]

#Mapa de color en escala de grises
color = [str(item/255.) for item in y]

#Definimos los puntos sombreados en base a una tercera variable
plt.scatter(x, y, s=850, c=color)

plt.show()
```

Cómo bombinar 4 o más gráficas en una sola figura?<br>
**Respuesta:** Para mostrar varios gráficos en una sola figura o ventana, podemos usar la función subplot(), indicando entre paréntesis un número con tres dígitos. El primer dígito indica el número de filas en los que se dividirá la figura, el segundo el número de columnas y el tercero se refiere al gráfico con el que estamos trabajando en ese momento.

```python
# Minimo comun multiplo de dos numeros
# Funcion que recibe mis dos numeros positivos
def mcm(x, y):
    if x > y:
        z = x
    else:
        z = y

    while(True):
        if((z % x == 0) and (z % y == 0)):
            mcm = z
            break
        z += 1

    return mcm


print(mcm(20, 8))
print(mcm(15, 17))
```
